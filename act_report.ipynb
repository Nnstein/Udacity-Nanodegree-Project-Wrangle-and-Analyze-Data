## DOGS ARE JUST SO CUTE.

#### This dataset was given to us by weratdogs,  some parts were gotten from the twitter API and the other parts was downloaded. The dataset is about the dogs that were twitted out by dog owners for them to be rated by weratedogs just for fun. 

#### The twitter_archive_master_csv file that we'll be analyzing has been gathered, thoroughly accerssed and cleaned for this analysis. This dataset contains 320 rows and 24 columns. Below is a description of the rows

### Column Description

`tweet_id`: contains the unique id for each tweet in the dataset

`timestamp`: contains the timestamp for each tweet i.ethe exact time each tweet was put out

`source`: the device or platform from were these tweets originated from

`tesxt`: contains the words and sentences that were put out in the tweet

`rating_numerator`: contains the numerator score for each dog picture that was put out

`rating_denominator`: this column contain the value '10' because according to weratedogs, all dogs are so cute that their ratings are from 10 and above

`name`: contains the name for each dogs whose photos were tweeted

`dog_stage`: shows the stage the dogs are in. it can be a 'floofer', 'puppo', 'pupper', and doggo', or a double combination of any of the satages like 'floofer puppo' 

`text_range`: this the length of the text in the text column

`favorite_count`: contains the amount of times each tweet was favorited

`favorited`: a boolean column that shows True if the tweet was favorited by weratedogs, and False if it was not favorited

`lang`: shows the language acronym each tweet was put out in

`retweet_count`: shows the number of times each tweet was retweeted

`jpg_url`: shows the link to the dog photos in each tweet

`img_num`: contains the number of photos for the dogs that was put out. it is either 1,2,3 or 4.

`p1`: this is the result of a machine learning algorithm that predicts the breed of dog based of the image that it was fed. this p1 column contains the 1st predicted breed name

`p1_conf`: THis shows the confidence level of the algorithm on the predicted breed on  a scale of 0-1.

`p1_dog`: Contains a boolean for the p1_conf. 

`p2`: this is the result of a machine learning algorithm that predicts the breed of dog based of the image that it was fed. this p2 column contains the 2nd predicted breed name

`p2_conf`: THis shows the confidence level of the algorithm on the predicted breed on  a scale of 0-1.

`p2_dog`: Contains a boolean for the p2_conf. 

`p3`: this is the result of a machine learning algorithm that predicts the breed of dog based of the image that it was fed. this p3 column contains the 3rd predicted breed name

`p3_conf`: THis shows the confidence level of the algorithm on the predicted breed on  a scale of 0-1.

`p3_dog`: Contains a boolean for the p3_conf.
